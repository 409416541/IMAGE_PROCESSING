```python
import numpy as np
from google.colab.patches import cv2_imshow

img = np.zeros([100, 512], dtype = 'uint8')

gray = 0

for i in range(100):
    for j in range(512):
        img[i, j] = gray
        if ((j + 1) % 2 == 0) & ((j + 1) > 0):
            gray += 1

print(img)
cv2_imshow(img)
```

```python
import numpy as np
import cv2
from google.colab import drive
from google.colab.patches import cv2_imshow

drive.mount('/content/drive')

imagePath = "/content/drive/My Drive/Colab Notebooks/image_processing/"

img = np.zeros([512, 512], dtype = 'uint8')

for i in range(8):
    for j in range(8):
        if((i + j) % 2 == 0):
            gray = 0
        else:
            gray = 255
        
        for l in range(64):
            for k in range(64):
                img[(i * 64) + l, (j * 64) + k] = gray

cv2.imwrite(imagePath+'chessboard.bmp', img)
print(img)
cv2_imshow(img)
```
