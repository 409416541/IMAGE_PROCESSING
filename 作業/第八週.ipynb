```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
from google.colab import drive
from google.colab.patches import cv2_imshow

drive.mount('/content/drive')
img = cv2.imread("/content/drive/My Drive/Colab Notebooks/image_processing/img_for_ch5.jpg", -1)
print(img.shape)

newimg = cv2.bilateralFilter(img, 5, 100, 100)

plt.figure(figsize = (15, 10.05))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('original', fontsize = 15, color = 'r')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(newimg, cv2.COLOR_BGR2RGB))
plt.title('Bilaterial Filtering', fontsize = 15, color = 'r')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def series(N, t):
    x = np.zeros(1000)
    for n in range(1, N + 1):
        x += (2 / (n * np.pi)) * ((np.sin(n * np.pi / 2)) - (np.sin(n * np.pi) - np.sin(n * np.pi / 2))) * np.cos(n * np.pi * t)
    return x

def subplot(plotindex, t, x, xlabel, ylabel):
    plt.subplot(plotindex)
    plt.plot(t, x)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

t = np.linspace(-2, 2, 1000)
plt.figure(figsize = (15, 10))

N = 2
x = series(N, t)
subplot(231, t, x, 't(second)', 'Amplotude' + str(N))
N = 5
x = series(N, t)
subplot(232, t, x, 't(second)', 'Amplotude' + str(N))
N = 10
x = series(N, t)
subplot(233, t, x, 't(second)', 'Amplotude' + str(N))
N = 50
x = series(N ,t)
subplot(234, t, x, 't(second)', 'Amplotude' + str(N))
N = 100
x = series(N, t)
subplot(235, t, x, 't(second)', 'Amplotude' + str(N))
N = 200
x = series(N, t)
subplot(236, t, x, 't(second)', 'Amplotude' + str(N))

plt.tight_layout()
plt.show()
```
